```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('../class_home/face.jpg',1)
img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
#load harr cascode classifier
face_detection = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
face=face_detection.detectMultiScale(gray,1.3,5)
face
```

```python
for(x,y,w,h) in face:
    images=cv2.rectangle(gray,(x,y),(x+w,y+h),(0,255,0),3)
plt.imshow(images,cmap='gray')
```
