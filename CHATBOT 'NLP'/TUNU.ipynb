```python
import random
import speech_recognition as Sr
from gtts import gTTS
import os
import time
#INPUT VOCE COMMAND

 

def voiceinput():
    r=Sr.Recognizer()
    with Sr.Microphone() as source:
        print('\nspeak something:')
        audio=r.listen(source)
        try:
            text=r.recognize_google(audio)
            you_said='{}'.format(text)
            print(you_said)
            return you_said
        except:
            print('No sound detected')
            #break
            
text = voiceinput()

user = [
        ['hi', 'hello'],
        ['how are you', 'how are you doing today'],
        ['thank you', 'thanks']
        ]


 

reply = [
    ['hello', 'good day to you'],
    ['I am good', 'It is a nice day today'],
    ['You are welcome']
    ]

 

while True:
    if text == "bye":
        #print("Have a good day ahead!!")
        language='en'
        output=gTTS(text='have a good day ahead',lang=language,slow=False)
        output.save('output1.mp3')
        os.system('start output1.mp3')
        time.sleep(3)
        os.remove('output1.mp3')
        break
    elif text == "":
        #print("You didn't type anything!!")
        language='en'
        output=gTTS(text='you havent said anything',lang=language,slow=False)
        output.save('output2.mp3')
        os.system('start output2.mp3')
        time.sleep(3)
        os.remove('output2.mp3')
        voiceinput()

 

    else:
        for i in range(len(user)):
            if text in user[i]:
               # print(random.choice(reply[i]))
                language='en'
                output=gTTS(text=random.choice(reply[i]),lang=language,slow=False)
                output.save('output3.mp3')
                os.system('start output3.mp3')
                time.sleep(3)
                os.remove('output3.mp3')
                text=voiceinput()
                
                break
        else:
            #text!=user
            language='en'
            output=gTTS('sorry i did not get it',lang=language,slow=False)
            output.save('output4.mp3')
            os.system('start output4.mp3')
            time.sleep(3)
            os.remove('output4.mp3')
            text=voiceinput()
            
            
```
